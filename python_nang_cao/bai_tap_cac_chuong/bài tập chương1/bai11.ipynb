```python
import math

class TamGiac:
    def __init__(self, a=1, b=1, c=1):
        self.a = a
        self.b = b
        self.c = c

    def nhap(self):
        while True:
            self.a = float(input("Nhap canh a: "))
            self.b = float(input("Nhap canh b: "))
            self.c = float(input("Nhap canh c: "))
            if self.la_tam_giac():
                break
            else:
                print("Ba canh khong tao thanh tam giac. Nhap lai!")

    def la_tam_giac(self):
        return (
            self.a + self.b > self.c and
            self.a + self.c > self.b and
            self.b + self.c > self.a
        )

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        p = self.chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

    def hien_thi(self):
        print(f"Tam giac co cac canh: a = {self.a}, b = {self.b}, c = {self.c}")
        print(f"Chu vi: {self.chu_vi()}")
        print(f"Dien tich: {self.dien_tich():.2f}")
class TamGiacVuong(TamGiac):
    def __init__(self):
        super().__init__()

    def nhap(self):
        while True:
            print("Nhap tam giac vuong:")
            super().nhap()
            if self.la_tam_giac_vuong():
                break
            else:
                print("Day khong phai tam giac vuong. Nhap lai!")

    def la_tam_giac_vuong(self):
        canh = sorted([self.a, self.b, self.c])
        return math.isclose(canh[0]**2 + canh[1]**2, canh[2]**2)
class TamGiacCan(TamGiac):
    def __init__(self):
        super().__init__()

    def nhap(self):
        while True:
            print("Nhap tam giac can:")
            super().nhap()
            if self.la_tam_giac_can():
                break
            else:
                print("Day khong phai tam giac can. Nhap lai!")

    def la_tam_giac_can(self):
        return (
            self.a == self.b or
            self.a == self.c or
            self.b == self.c
        )
class TamGiacDeu(TamGiacCan):
    def __init__(self):
        super().__init__()

    def nhap(self):
        while True:
            print("Nhap tam giac deu:")
            self.a = float(input("Nhap do dai canh: "))
            self.b = self.a
            self.c = self.a
            if self.la_tam_giac_deu():
                break

    def la_tam_giac_deu(self):
        return self.a == self.b == self.c
def main():
    print("\n=== TAM GIAC THUONG ===")
    tg = TamGiac()
    tg.nhap()
    tg.hien_thi()

    print("\n=== TAM GIAC VUONG ===")
    tgv = TamGiacVuong()
    tgv.nhap()
    tgv.hien_thi()

    print("\n=== TAM GIAC CAN ===")
    tgc = TamGiacCan()
    tgc.nhap()
    tgc.hien_thi()

    print("\n=== TAM GIAC DEU ===")
    tgd = TamGiacDeu()
    tgd.nhap()
    tgd.hien_thi()
main()

```
