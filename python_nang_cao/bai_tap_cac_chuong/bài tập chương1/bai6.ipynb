```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.data = []
        self.top = -1

    def __del__(self):
        print("Giai phong bo nho ngan xep")
        del self.data

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.capacity - 1

    def push(self, value):
        if self.isFull():
            print("Ngan xep da day, khong the push.")
        else:
            self.data.append(float(value))
            self.top += 1

    def pop(self):
        if self.isEmpty():
            print("Ngan xep rong, khong the pop.")
            return None
        else:
            value = self.data.pop()
            self.top -= 1
            return value

    def count(self):
        return self.top + 1

    def print_stack(self):
        if self.isEmpty():
            print("Ngan xep rong.")
        else:
            print("Noi dung ngan xep (tu dinh xuong day):")
            for i in range(self.top, -1, -1):
                print(f"  {self.data[i]}")
stack = Stack(4)
stack.push(1.1)
stack.push(2.2)
stack.push(3.3)

print(f"So phan tu hien co:{stack.count()}")

stack.print_stack()

stack.pop()

print("Sau khi pop:")
stack.print_stack()
```
