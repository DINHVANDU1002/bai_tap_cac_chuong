```python
class PS:
    def __init__(self):
        self.tu_so = 0
        self.mau_so = 1  

    def nhap_phan_so(self):
        self.tu_so = int(input("Nhap tu so: "))
        while True:
            self.mau_so = int(input("Nhap mau so (khac 0): "))
            if self.mau_so != 0:
                break
            else:
                print("Mau so phai khac 0, vui long nhap lai.")

    def kiem_tra_hop_le(self):
        return self.mau_so != 0

    def in_phan_so(self):
        if self.kiem_tra_hop_le():
            print(f"Phan so: {self.tu_so}/{self.mau_so}")
        else:
            print("Phan so khong hop le (mau so = 0).")
ps = PS()
ps.nhap_phan_so()
ps.in_phan_so()
```
